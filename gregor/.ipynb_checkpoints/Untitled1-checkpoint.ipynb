```python
def convert(h):
    #return 2**16-1
    normalized = h/1024
    to16bit=(2**16-1)*normalized
    rounded = int(to16bit)
    if rounded>=2**16 or rounded<0:
        print("error")
    return rounded

x1, y1, x2, y2 = (45.978571, 14.395671, 46.140379, 14.627671)
#img=[[0. for _ in range(1081)] for _ in range(1081)]
img=np.zeros((1081,1081), dtype=np.uint16)
done=np.zeros((1081,1081), dtype=np.uint16)

for i in range(len(ls)):
    x=ls[i][0]
    y=ls[i][1]
    px=1080-int(1081*(x-x1)/(x2-x1))
    py=int(1081*(y-y1)/(y2-y1))
    
    img[px][py] = convert(ls[i][2])
#     if px>200 and px<400 and py>100 and py+px<500:
#         img[px][py]=2**16-1
#     else:
#         img[px][py]=0
    #done[px][py] = 1
```
